# Seaweed_Detection
이 프로젝트는 2024 DATA·AI 분석 경진대회 참가를 위해 제작하였습니다. 사전 학습된 YOLO 모델(best.pt)을 활용하여 김(Seaweed) 객체를 감지하고 분류합니다.  

## 프로젝트에 사용된 기술 및 버전
### | Language
* Python `3.11.7`
* torch `2.4.1+cu121`
* ultralytics `8.2.99`
* opencv-python `4.10.0.84`
### | System
* Windows 10
* Driver Version: 531.14
* CUDA Version: 12.1

## 종속성 설치
아래 단계에 따라 환경을 설정하고 모델을 실행합니다. requirements.txt 파일을 사용하여 환경을 설정할 수 있습니다.
* python 3.11.7 가상환경 생성
```python
pip install -r requirements.txt
```

## 디렉터리 구성
```
 ┣ data

* jsontotxt.py : json 형식의 라벨링 데이터를 yolo 학습에 맞게 txt 형식으로 변경하는 코드
* predict.py : 테스트셋 평가 코드, results 폴더에 visualization 폴더와 json 파일 생성
* train.py : 지정된 하이퍼파라미터를 사용한 모델 학습 코드
* trainWB.py : WandB툴을 활용한 하이퍼파라미터 스윕 학습 코드
* 모든 python 파일 실행은 notebooks 폴더 밖에서 실행, "cd .."으로 폴더 밖으로 나온 뒤 수행

## 모델 학습

```
```python
python src/trainWB.py
```

## 모델 평가
validation셋을 사용한 정량적 모델 평가 수행
```python
yolo detect val model=./model/best.pt data=./data/data.yaml imgsz=1024 project=results_dir name=val_run
```

## 예측 수행
사전 학습 모델(best.pt)을 활용하여 예측을 수행
```python

```
